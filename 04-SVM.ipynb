```python
import mglearn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris
```

```python
X = iris.data[50:, 2:]
y = iris.target[50:] - 1
```

```python
X
```

```python
y
```

```python
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(['versicolor', 'virginica'], loc='best')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=0)
```

```python
svm = LinearSVC().fit(X_train, y_train)
```

```python
def plot_separator(model):
    plt.figure(figsize=(10, 6))
    mglearn.plots.plot_2d_separator(model, X)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
    plt.xlabel('petal length')
    plt.ylabel('petal width')
    plt.xlim(2.8, 7.0)
    plt.ylim(0.8, 2.6)
    plt.show()
```

```python
plot_separator(svm)
```

```python
svm_100 = LinearSVC(C=100).fit(X_train, y_train)
plot_separator(svm_100)
```

```python
print('score on training set: {:.2f}'.format(svm.score(X_train, y_train)))
print('score on test set: {:.2f}'.format(svm.score(X_test, y_test)))
```

```python
print('score on training set: {:.2f}'.format(svm_100.score(X_train, y_train)))
print('score on test set: {:.2f}'.format(svm_100.score(X_test, y_test)))
```

```python
svm_001 = LinearSVC(C=0.01).fit(X_train, y_train)
plot_separator(svm_001)
```

```python
print('score on training set: {:.2f}'.format(svm_001.score(X_train, y_train)))
print('score on test set: {:.2f}'.format(svm_001.score(X_test, y_test)))
```

```python
import numpy as np
from sklearn.datasets import make_moons
```

```python
moons = make_moons(n_samples=200, noise=0.1, random_state=0)
moons
```

```python
X = moons[0]
y = moons[1]
```

```python
plt.figure(figsize=(12, 8))
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=0)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
lin_svm = LinearSVC().fit(X_train_scaled, y_train)

plt.figure(figsize=(12, 8))
mglearn.plots.plot_2d_separator(lin_svm, X)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel("$x_0$", fontsize=20)
plt.ylabel("$x_1$", fontsize=20)
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=3)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.fit_transform(X_test)

print(X_train_poly.shape)
print(X_train_poly)
```

```python
poly.get_feature_names()
```

```python
X_train_poly_scaled = scaler.fit_transform(X_train_poly)
X_test_poly_scaled = scaler.fit_transform(X_test_poly)

lin_svm = LinearSVC().fit(X_train_poly_scaled, y_train)
lin_svm.predict(X_test_poly_scaled) == y_test
```

```python
from sklearn.pipeline import Pipeline

poly_svm = Pipeline([
    ('poly', PolynomialFeatures(degree=3)),
    ('scaler', StandardScaler()),
    ('svm', LinearSVC())
])

poly_svm.fit(X, y)
```

```python
def plot_decision_function(model):
    _x0 = np.linspace(-1.5, 2.5, 100)
    _x1 = np.linspace(-1.0, 1.5, 100)
    x0, x1 = np.meshgrid(_x0, _x1)
    X = np.c_[x0.ravel(), x1.ravel()]
    y_pred = model.predict(X).reshape(x0.shape)
    y_decision = model.decision_function(X).reshape(x0.shape)
    plt.contourf(x0, x1, y_pred, cmap=plt.cm.brg, alpha=0.2)
    plt.contourf(x0, x1, y_decision, levels=[y_decision.min(), 0, y_decision.max()], alpha=0.3)
```

```python
def plot_dataset(X, y):
    plt.plot(X[:, 0][y==0], X[:, 1][y==0], "bo", ms=15)
    plt.plot(X[:, 0][y==1], X[:, 1][y==1], "r^", ms=15)
    plt.xlabel("$x_1$", fontsize=20)
    plt.ylabel("$x_2$", fontsize=20, rotation=0)
```

```python
plt.figure(figsize=(12, 8))
plot_decision_function(poly_svm)
plot_dataset(X, y)
plt.show()
```

```python
from sklearn.svm import SVC
```

```python
kernel_svm = Pipeline([
    ('scaler', StandardScaler()),
    ('svm', SVC(kernel='poly', degree=3, coef0=1))
])

kernel_svm.fit(X, y)
```

```python
plt.figure(figsize=(12, 8))
plot_decision_function(kernel_svm)
plot_dataset(X, y)
plt.show()
```

```python
plt.figure(figsize=(20, 15))

for i, degree in enumerate([2, 3, 5, 10]):
    poly_kernel_svm = Pipeline([
        ('scaler', StandardScaler()),
        ('svm', SVC(kernel='poly', degree=degree, coef0=1))
    ])
    poly_kernel_svm.fit(X, y)
    
    plt.subplot(221 + i)
    plot_decision_function(poly_kernel_svm)
    plot_dataset(X, y)
    plt.title("d = {}".format(degree), fontsize=30)

plt.show()
```
